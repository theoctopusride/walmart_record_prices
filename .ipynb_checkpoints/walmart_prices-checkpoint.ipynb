```python
def create_sorts():
    sorts = []
    sorts.append("&sort=price_low")
    sorts.append("&sort=price_high")
    sorts.append("&sort=rating_high")
    sorts.append("&sort=new")
    return sorts
```

```python
#walmart's website has multiple different categories for different genres/types of records,
#but they only display 1000 per genre. We split the genre categories into those that have more
#than 1000 records and those that have less than 1000:
def create_categories():
    categories_greater_1000 = []
    #rock records > 1000 total:
    categories_greater_1000.append("?cat_id=4104_1205481_2844615")
    #jazz records > 1000 total:
    categories_greater_1000.append("?cat_id=4104_1205481_3708052")
    #pop records > 1000 total:
    categories_greater_1000.append("?cat_id=4104_1205481_6204731")
    #r&b and soul records > 1000 total:
    categories_greater_1000.append("?cat_id=4104_1205481_7409068")
    #hip hop records > 1000 total:
    categories_greater_1000.append("?cat_id=4104_1205481_7766072")
    categories_less_1000 = []
    #blues records < 1000 total:
    categories_less_1000.append("?cat_id=4104_1205481_3605866")
    #childrens records < 1000 total:
    categories_less_1000.append("?cat_id=4104_1205481_7687262")
    #country records < 1000 total:
    categories_less_1000.append("?cat_id=4104_1205481_4846487")
    #latin records < 1000 total:
    categories_less_1000.append("?cat_id=4104_1205481_1630455")
    #soundtrack records < 1000 total:
    categories_less_1000.append("?cat_id=4104_1205481_9901623")
    
    return categories_greater_1000, categories_less_1000
```

```python
def num_items_and_pages_per_filter(ww1, ww2, headers):
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:68.0)'}
    #main vinyl page link split into two so that page numbers can be added in the middle:
    ww1 = "https://www.walmart.com/browse/music/vinyl/4104_1205481?page="
    ww2 = "#searchProductResult"
    wally_world_vinyl_page = ww1 + "1" + ww2
    page = requests.get(wally_world_vinyl_page, headers=headers)
    soup = BeautifulSoup(page.text, 'html.parser')
    results = soup.find("div", class_="result-summary-container")
    results = str(results.text)
    results_beg = results.find(" of ")+4
    results = results[results_beg:]
    if "," and "+" in results:
        comma = results.find(",")
        plus = results.find("+")
        results = results[:comma] + results[comma+1:plus] + results[plus+1:]
    results_end = results.find(" products")
    results = results[:results_end]
    results = int(results)
    page_results = math.ceil(results/40)
    return results, page_results
```

```python
def check_ww_page(soup, num_pages, releases_check):
    releases = releases_check
    script = soup.find_all("script")
    artist_title_info = str(script[25])
    while "title\":" in artist_title_info:
        product_id_beg = artist_title_info.find("productId\":")+12
        artist_title_info = artist_title_info[product_id_beg:]
        product_id_end = artist_title_info.find("\",\"")
        product_id = artist_title_info[:product_id_end]
        us_id_beg = artist_title_info.find("usItemId\":")+11
        artist_title_info = artist_title_info[us_id_beg:]
        us_id_end = artist_title_info.find("\",\"")
        us_id = artist_title_info[:us_id_end]
        title_name_beg = artist_title_info.find("title\":")+8
        if title_name_beg > 0:
            artist_title_info = artist_title_info[title_name_beg:]
            title_name_end = artist_title_info.find("\",\"");
            release_info = []
            title_name = artist_title_info[:title_name_end]
            if "Vinyl)" in title_name:
                group_beg = artist_title_info.find("seeAllName\":")+13
                artist_title_info = artist_title_info[group_beg:]
                group_end = artist_title_info.find("\",\"")
                group = artist_title_info[:group_end]
                price_beg = artist_title_info.find("offerPrice\":")+12
                artist_title_info = artist_title_info[price_beg:]
                price_end = artist_title_info.find(",\"")
                price = f'{float(artist_title_info[:price_end]):.2f}'
                artist_name_beg = artist_title_info.find("artistName\":")+13
                artist_title_info = artist_title_info[artist_name_beg:]
                artist_name_end = artist_title_info.find("\",\"")
                artist_name = artist_title_info[:artist_name_end]
                release_info.append(artist_name)
                artist_title_info = artist_title_info[artist_name_end:]
                quantity_beg = artist_title_info.find("quantity\":")+10
                artist_title_info = artist_title_info[quantity_beg:]
                quantity_end = artist_title_info.find(",\"")
                quantity = artist_title_info[:quantity_end]
                release_info.append(title_name)
                release_info.append(product_id)
                release_info.append(group)
                release_info.append(price)
                release_info.append(quantity)
                releases[us_id] = release_info
    return releases;
```
